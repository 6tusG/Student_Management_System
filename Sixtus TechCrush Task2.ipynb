### Student Result Management System

To build a Student Result Management System that will help teachers manage student records efficiently.

### Add Student Task

To build a function that asks for a student's name and score, and stores it in a dictionary.

```python
# This dictionary will hold all student records.
# Key = student name, Value = score

students = {}
```

```python
# Function to calculate grade based on score

def calculate_grade(score):
    if score >= 70 and score <= 100:
        return "A"
    elif score >= 60 and score <= 69:
        return "B"
    elif score >= 50 and score <= 59:
        return "C"
    elif score >= 45 and score <= 49:
        return "D"
    elif score >= 0 and score <= 44:
        return "F"
    else:
        return "Invalid Score"
```

```python
# Add Student Function

def add_student():
    print(" ADD STUDENT ")
    
    name = input("Enter student name: ")
    
    if name in students:
        print("Error: Student already exists.")
        return
    
    while True:
        try:
            score = float(input("Enter student score (0-100): "))
            if 0 <= score <= 100:
                break
            else:
                print("Error: Score must be between 0 and 100.")
        except ValueError:
            print("Error: Invalid input. Please enter a number.")
    
    students[name] = score
    print("Student added successfully.")
```

### View All Students Task

To build a function that displays All Students' name, score, and grade

```python
# View All Students Function

def view_students():
    print("View ALL Students")
    
    if len(students) == 0:
        print("No students found.")
        return
    
    print("Name         Score    Grade")
    
    for name in students:
        score = students[name]
        grade = calculate_grade(score)
        print(f"{name}         {score}      {grade}")
```

### Update Student Score Task

This function will allow teachers to update the score of an existing student.


```python
def update_score():
    print("Update Student Score")
    
    name = input("Enter student name: ")
    
    if name not in students:
        print("Error: Student not found.")
        return
    
    print(f"Current score for {name}: {students[name]}")
    
    while True:
        try:
            new_score = float(input("Enter new score (0-100): "))
            if 0 <= new_score <= 100:
                break
            else:
                print("Error: Score must be between 0 and 100.")
        except ValueError:
            print("Error: Invalid input. Please enter a number.")
    
    students[name] = new_score
    print("Score updated successfully.")
```

### Search for a Student Task

This function will allow users to search for a student by name.

```python
def search_student():
    print("Search Student")
    
    name = input("Enter student name to search: ")
    
    if name in students:
        score = students[name]
        grade = calculate_grade(score)
        print("Student found!")
        print(f"Name: {name}")
        print(f"Score: {score}")
        print(f"Grade: {grade}")
    else:
        print("Error: Student not found.")
```

### Display Class Statistics Task

This function will display summary statistics about all students in the system.

It will show:
1. Total number of students
2. Highest score
3. Lowest score
4. Average score (rounded to 2 decimal places)

```python
def class_statistics():
    print("Class Statistics")
    
    if len(students) == 0:
        print("No students found.")
        return
    
    scores = list(students.values())
    
    total_students = len(students)
    highest = max(scores)
    lowest = min(scores)
    average = sum(scores) / len(scores)
    
    print(f"Total Students: {total_students}")
    print(f"Highest Score: {highest}")
    print(f"Lowest Score: {lowest}")
    print(f"Average Score: {average:.2f}")
```

### Menu System

This is the main menu that will run the entire Student Result Management System.

```python
def main_menu_students():
    print("STUDENT RESULT MANAGEMENT SYSTEM")
    print("1. Add Student")
    print("2. View All Students")
    print("3. Update Student Score")
    print("4. Search for a Student")
    print("5. Class Statistics")
    print("6. Exit")
```

```python
def run_student_system():
    while True:
        main_menu_students()
        
        choice = input("Enter your choice (1-6): ")
        
        if choice == "1":
            add_student()
        elif choice == "2":
            view_students()
        elif choice == "3":
            update_score()
        elif choice == "4":
            search_student()
        elif choice == "5":
            class_statistics()
        elif choice == "6":
            print("Exiting Student System. Goodbye!")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 6.")
        
        print()
```

## FreshMart Stores 

### INVENTORY MANAGEMENT SYSTEM

To build an Inventory Management System that will help FreshMart manage its inventory more efficiently.

```python
# This dictionary will hold all inventory items.
# Key = product name, Value = quantity
inventory = {}
```

### Add Product Task

This function will allow users to add new products to the inventory.

```python
#add product function

def add_product():
    print("Add Product")
    
    name = input("Enter product name: ")
    
    if name in inventory:
        print("Error: Product already exists.")
        return
    
    while True:
        try:
            quantity = int(input("Enter quantity: "))
            if quantity >= 0:
                break
            else:
                print("Error: Quantity cannot be negative.")
        except ValueError:
            print("Error: Invalid input. Please enter a number.")
    
    inventory[name] = quantity
    print("Product added successfully.")
```

### View Inventory Task

This function will display all products and their quantities.

```python
#view_inventory function

def view_inventory():
    print("View Inventory")
    
    if len(inventory) == 0:
        print("Inventory is empty.")
        return
    
    print("Product         Quantity")
    
    for name in inventory:
        quantity = inventory[name]
        print(f"{name}         {quantity}")
```

### Update Product Quantity Task

This function will allow users to modify the quantity of an existing product.

```python
#update_quantity function

def update_quantity():
    print("Update Quantity")
    
    name = input("Enter product name: ")
    
    if name not in inventory:
        print("Error: Product not found.")
        return
    
    print(f"Current quantity for {name}: {inventory[name]}")
    
    while True:
        try:
            new_quantity = int(input("Enter new quantity: "))
            if new_quantity >= 0:
                break
            else:
                print("Error: Quantity cannot be negative.")
        except ValueError:
            print("Error: Invalid input. Please enter a number.")
    
    inventory[name] = new_quantity
    print("Quantity updated successfully.")
```

### Search Product Task

This function will allow users to search for a product by name.

```python
#search_product function

def search_product():
    print("Search Product")
    
    name = input("Enter product name to search: ")
    
    if name in inventory:
        quantity = inventory[name]
        print("Product found!")
        print(f"Name: {name}")
        print(f"Quantity: {quantity}")
    else:
        print("Error: Product not found.")
```

### Remove Product Task

This function will allow users to remove a product from the inventory.

```python
#remove_product function

def remove_product():
    print("Remove Product")
    
    name = input("Enter product name to remove: ")
    
    if name not in inventory:
        print("Error: Product not found.")
        return
    
    confirm = input(f"Are you sure you want to remove '{name}'? (yes/no): ")
    
    if confirm == "yes":
        del inventory[name]
        print("Product removed successfully.")
    else:
        print("Removal cancelled.")
```

### Low Stock Report Task

This function will display all products with quantity less than 5.

```python
def low_stock_report():
    print("Low Stock Report")
    
    low_stock_items = {}
    
    for name in inventory:
        quantity = inventory[name]
        if quantity < 5:
            low_stock_items[name] = quantity
    
    if len(low_stock_items) == 0:
        print("All products have sufficient stock (5 or more).")
        return
    
    print("Product         Quantity")
    
    for name in low_stock_items:
        quantity = low_stock_items[name]
        print(f"{name}         {quantity}")
```

### Inventory Statistics Task

This function will display summary statistics about the inventory.

It will show:
1. Total number of products
2. Product with the highest quantity
3. Product with the lowest quantity
4. Total stock available (sum of all quantities)

```python
def inventory_statistics():
    print("Inventory Statistics")
    
    if len(inventory) == 0:
        print("Inventory is empty.")
        return
    
    quantities = list(inventory.values())
    
    total_products = len(inventory)
    total_stock = sum(quantities)
    highest_qty = max(quantities)
    lowest_qty = min(quantities)
    
    # Find product names for highest and lowest quantities
    highest_product = ""
    lowest_product = ""
    
    for name in inventory:
        if inventory[name] == highest_qty:
            highest_product = name
        if inventory[name] == lowest_qty:
            lowest_product = name
    
    print(f"Total Products: {total_products}")
    print(f"Total Stock Available: {total_stock}")
    print(f"Product with Highest Quantity: {highest_product} ({highest_qty})")
    print(f"Product with Lowest Quantity: {lowest_product} ({lowest_qty})")
```

### Menu System

This is the main menu that will run the entire Inventory Management System.

```python
def main_menu_inventory():
    print("INVENTORY MANAGEMENT SYSTEM")
    print("1. Add Product")
    print("2. View Inventory")
    print("3. Update Quantity")
    print("4. Search Product")
    print("5. Remove Product")
    print("6. Low Stock Report")
    print("7. Inventory Statistics")
    print("8. Exit")
```

```python
#run_inventory_system function

def run_inventory_system():
    while True:
        main_menu_inventory()
        
        choice = input("Enter your choice (1-8): ")
        
        if choice == "1":
            add_product()
        elif choice == "2":
            view_inventory()
        elif choice == "3":
            update_quantity()
        elif choice == "4":
            search_product()
        elif choice == "5":
            remove_product()
        elif choice == "6":
            low_stock_report()
        elif choice == "7":
            inventory_statistics()
        elif choice == "8":
            print("Exiting Inventory System. Goodbye!")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 8.")
        
        print()
```
